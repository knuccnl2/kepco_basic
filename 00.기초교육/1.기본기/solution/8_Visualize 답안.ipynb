### 실습예제 4
위의 코드를 참고하여 9개의 랜덤한 히스토그램을 그리시오.

```python
fig, axes = plt.subplots(3, 3, sharex=True, sharey=True)
for i in range(3):
    for j in range(3):
        axes[i, j].hist(randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

### 실습예제6
위의 그래프의 범례를 오른쪽 아래 위치로 이동시키십시오.(힌트 : location 위치에 아무 글씨나 집어넣으면 힌트가 나옵니다.)

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')

ax.legend(loc='lower right')
plt.show()
```

### 실습예제10
위 그래프를 가로축 중심의 그래프로 그리시오.

```python
tips = pd.read_csv('data/tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
print(party_counts)
# 2~5명의 파티 데이터만 추출
ptt=party_counts.ix[:, 2:5]
print(ptt)
```

```python
party_pcts = ptt.div(ptt.sum(1).astype(float), axis=0)
print(party_pcts)
party_pcts.plot(kind='barh', stacked=True)
plt.show()
```

```python
party_pcts = ptt.div(ptt.sum(1).astype(float), axis=0)
print(party_pcts)

party_pcts.plot(kind='bar', stacked=True)
plt.show()
```

### 실습예제11
위의 그래프의 막대 부분의 색을 진하게 하고 점선으로 된 커널 밀도 추정 그래프를 실선으로 표시하시오.

```python
comp1 = np.random.normal(0, 1, size=200)  # N(0, 1)
comp2 = np.random.normal(10, 2, size=200)  # N(10, 4)
values = Series(np.concatenate([comp1, comp2]))
values.hist(bins=100, alpha=0.8, color='k', normed=True)
values.plot(kind='kde', style='k-')
plt.show()
```

### 실습예제12
위의 Scatter plot의 x축을 trans_data의 cpi, y축을 tbilrate로 한 scatter plot을 그리고 제목을 그에 맞게 바꾸시오.

```python
macro = pd.read_csv('data/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna() #dropna는 nan 즉 0의 값을 뺀다는 의미. diff()는 미분하는 함수, log()는 log함수를 취하는 함수.
```

```python
plt.scatter(trans_data['cpi'], trans_data['tbilrate'])
plt.title('Changes in log %s vs. log %s' % ('cpi', 'tbilrate'))
plt.show()
```
